```python
# Set style for all plots
def set_style():
    plt.style.use('seaborn-v0_8')
    sns.set_palette("husl")
    plt.rcParams['figure.figsize'] = (12, 6)
    plt.rcParams['font.size'] = 12
```

```python
# Visualization 1: Correlation Heatmap
def plot_correlation_heatmap(df):
    plt.figure(figsize=(10, 8))
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Between Metrics')
    plt.show()
```

```python
# Visualization 2: CGPA vs Screen Time
def plot_cgpa_vs_screen_time(df):
    sns.scatterplot(data=df, x='Screen Time', y='CGPA', hue='Gender')
    plt.title('CGPA vs Screen Time (Grouped by Gender)')
    plt.show()
```

```python

# Visualization 3: Sleep Duration by Age Groups
def plot_sleep_by_age(df):
    df['Age_Group'] = pd.qcut(df['Age'], q=4, labels=['18-20', '21-22', '23-24', '25+'])
    sns.boxplot(data=df, x='Age_Group', y='Sleep Duration')
    plt.title('Sleep Duration Across Age Groups')
    plt.show()
```

```python
# Visualization 4: Happiness Index by CGPA
def plot_happiness_vs_cgpa(df):
    sns.boxplot(data=df, x='Happiness Index', y='CGPA', hue='Gender')
    plt.title('Happiness Index vs CGPA (Gender Comparison)')
    plt.show()
```

```python
# Visualization 5: Social Media Usage by Age
def plot_social_media_by_age(df):
    sns.scatterplot(data=df, x='Age', y='Social Media Usage', hue='Gender')
    plt.title('Social Media Usage by Age')
    plt.show()
```

```python
# Visualization 6: Stress Index Distribution
def plot_stress_distribution(df):
    sns.histplot(df['Stress Index'], kde=True, color='blue')
    plt.title('Stress Index Distribution')
    plt.show()
```

```python

# Visualization 7: CGPA by Attendance
def plot_cgpa_vs_attendance(df):
    sns.lineplot(data=df, x='Attendance', y='CGPA', hue='Gender')
    plt.title('CGPA vs Attendance')
    plt.show()
```

```python
# Visualization 8: Sleep Duration vs Stress Index
def plot_sleep_vs_stress(df):
    sns.scatterplot(data=df, x='Sleep Duration', y='Stress Index', hue='Gender')
    plt.title('Sleep Duration vs Stress Index')
    plt.show()
```

```python
# Visualization 9: Screen Time by Happiness Index
def plot_screen_by_happiness(df):
    sns.boxplot(data=df, x='Happiness Index', y='Screen Time')
    plt.title('Screen Time by Happiness Index')
    plt.show()
```

```python
# Visualization 10: Age Group vs CGPA
def plot_age_group_vs_cgpa(df):
    sns.boxplot(data=df, x='Age_Group', y='CGPA')
    plt.title('CGPA Across Age Groups')
    plt.show()
```

```python
# Generate all visualizations
def generate_all_plots(df):
    set_style()
    plot_correlation_heatmap(df)
    plot_cgpa_vs_screen_time(df)
    plot_sleep_by_age(df)
    plot_happiness_vs_cgpa(df)
    plot_social_media_by_age(df)
    plot_stress_distribution(df)
    plot_cgpa_vs_attendance(df)
    plot_sleep_vs_stress(df)
    plot_screen_by_happiness(df)
    plot_age_group_vs_cgpa(df)

if __name__ == "__main__":
    generate_all_plots(df)
```
